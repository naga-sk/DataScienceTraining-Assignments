```python
# Get specific rows of the dataframe by referencing the index range. Notice the zero indexing format, and the increment
# Note: use of <loc> is cleaner
df[20:30:2]
```

```python
# the iloc (integer location) method is the same as the index range method above
# Note: use of <loc> is cleaner
df.iloc[20:30]
```

```python
# the loc (locate) method references the labels instead of the index. 
# In contrast with the iloc function, the loc function will get all the labels referenced
df.loc[20:30]
```

```python
# Use of loc to print specific labels
df.loc[{20,30}]
```

```python
# Use of loc to print specific labels
df.loc[20:30][df.Age > 35]
```
